###**Handling Inconsistent Values in the Dependents Column**

```r
loan_data$Dependents <- as.character(loan_data$Dependents)
loan_data$Dependents[loan_data$Dependents == "3+"] <- "3"
loan_data$Dependents <- as.numeric(loan_data$Dependents)
```

### **Recoding Categorical Variables into Numeric Format**

```r
# Loop through all columns to convert categorical data into clean numerical formats
for (col in names(loan_data)) {

  if (is.character(loan_data[[col]]) || is.factor(loan_data[[col]])) {

 # skip unique identifier columns like Loan_ID and the Property_Area
if (col %in% c("Loan_ID", "Property_Area")) next

# here i am handling binary columns directly to 0 and 1
    unique_vals <- unique(na.omit(loan_data[[col]]))
    if (length(unique_vals) == 2) {
     loan_data[[col]] <- ifelse(loan_data[[col]] == unique_vals[1], 1, 0)

    } else {
      loan_data[[col]] <- as.factor(as.factor(loan_data[[col]]))
    }
  }
}
#Here I converted the categorical columns into numerical values so they can be analyzed more easily.
 # For columns with two categories, I encoded them as 0 and 1. For columns with more than two categories, I assigned numeric labels to each category.
 #I excluded Loan_ID because it is just an identifier and does not add any value to the analysis.
# This step makes the data more consistent and ready for further analysis and modeling.
```

## **Data cleaning and transformation:**

At this point I found some missing values in the dataset that needed to be fixed. I decided not to remove the rows containing missing values because that would lead to losing a considerable amount of data, which would not be helpful for the analysis. Instead, I handled the missing values using suitable imputation methods to preserve as much information as possible.

###**Filling Missing Values in Categorical Variables Using the Mode**

```r
#Grouping specific target columns into a vector
target_columns <- c("Gender", "Married", "Self_Employed", "Dependents", "Credit_History")

# Define a helper function to find the Mode of the column
get_mode <- function(v) {
  uniqv <- unique(na.omit(v))
  uniqv[which.max(tabulate(match(v, uniqv)))]
}

# a for loop to go over the chosen columns
for (col in target_columns) {

  #make sure the column exists in your dataset
  if (col %in% names(loan_data)) {

    # Calculate the mode for this specific column
    col_mode <- get_mode(loan_data[[col]])

    # Replace all NA values in this column with its mode
    loan_data[is.na(loan_data[[col]]), col] <- col_mode
  }
}
```

- In this step I used a custom Mode function to handle missing values in categorical columns. First I grouped the target columns  into a vector. Then I created a function called get_mode() that identifies the most frequently occurring value in each column. Using a for loop, I calculated the mode for every selected column and replaced all missing values with that mode.
- I chose this approach because these variables are categorical, and the mode is the most appropriate measure for representing the most common category. This helps preserve the original distribution of the data while keeping all records available for analysis instead of removing rows with missing values.

**I wanted to visually check the mode to make sure the missing values were replaced correctlys so I displayed the frequency table for each categorical column using table().This step helped validate the imputation process before moving on to the next stage of data preparation.**

### **Replace missing numeric values with median to avoid having outliers**

```r
loan_data$Loan_Amount_Term[is.na(loan_data$Loan_Amount_Term)] <- median(loan_data$Loan_Amount_Term,na.rm = TRUE)
loan_data$LoanAmount[is.na(loan_data$LoanAmount)] <- median(loan_data$LoanAmount, na.rm = TRUE)

sum(is.na(loan_data$Loan_Amount_Term))
sum(is.na(loan_data$LoanAmount))
```

Unlike the previous categorical variables, Loan_Amount_Term and LoanAmount are numerical variables. For this reason, I used median imputation instead of mode imputation to handle their missing values. The median is more suitable for numerical data because it is less affected by  outliers and provides a better representation of the central tendency of the data. ## **Data Analysis:**

### **Credit History vs Loan Status**

Here I used a cross-tabulation with row percentages to compare loan approval rates. This will help to identify what influences the loan approval decision.

```r
prop.table(
  table(
    Credit_History = ifelse(loan_data$Credit_History == 1,
                            "Good Credit History",
                            "Poor Credit History"),
    Loan_Status = ifelse(loan_data$Loan_Status == 1,
                         "Approved",
                         "Rejected")
  ),
  1
)
```

**Credit history has a strong impact on loan approval.**
- Applicants with a good credit history were approved much more frequently (79%), while applicants with a poor credit history were rejected in most cases (92%).

### **Education vs Loan Status**

```r
prop.table(
  table(
    Education = ifelse(loan_data$Education == 1, "Graduate", "Not Graduate"),
    Loan_Status = ifelse(loan_data$Loan_Status == 1, "Approved", "Rejected")
  ),
  1
)
```

**Education appears to have a moderate effect on loan approval.**
 - Graduates had a higher approval rate (70.8%) compared to non-graduates (61.2%).


### **Married vs Loan Status**

```r
prop.table(
  table(
    Married = ifelse(loan_data$Married == 1, "Yes", "No"),
    Loan_Status = ifelse(loan_data$Loan_Status == 1, "Approved", "Rejected")
  ),
  1
)
```

**Marital status does not appear to have a strong impact on loan approval.**
-  Both married and unmarried applicants had relatively high approval rates, although unmarried applicants were approved slightly more often (71.8%) than married applicants (62.9%).

### **Property Area vs Loan Status**

```r
prop.table(
  table(
    loan_data$Property_Area,
    loan_data$Loan_Status
  ),
  1
)
```

**Property area appears to influence loan approval rates.**
 - Applicants from semi-urban areas had the highest approval rate (76.8%), followed by urban areas (65.8%), while rural areas had the lowest approval rate (61.5%). This suggests that applicants living in semi-urban areas were more likely to have their loans approved.
